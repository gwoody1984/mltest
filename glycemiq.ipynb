```python
import pandas as pd
import numpy as np
import math

data = pd.read_csv('c:/users/sunil/downloads/bgdata.csv')
data.head(100)
```

```python
data['diff5'] = np.log(data['bg']) - np.log(data['bg-5'])
data['diff10'] = np.log(data['bg-5']) - np.log(data['bg-10'])
data['diff15'] = (np.log(data['bg-10']) - np.log(data['bg-15'])) / 2
data['diff20'] = (np.log(data['bg-15']) - np.log(data['bg-20'])) / 4
data['diff25'] = (np.log(data['bg-20']) - np.log(data['bg-25'])) / 8
data['diff30'] = (np.log(data['bg-25']) - np.log(data['bg-30'])) / 16
data['difflabel'] = np.log(data['label']) - np.log(data['bg'])
cols = ['label','bg','difflabel','diff5','diff10','diff15','diff20','diff25','diff30','glycemicindex','calories','carbs','fiber','sugar','basal_insulin','bolus_insulin']
x_full = data[cols]
x_full.head(20)
```

```python
ix = x_full.isnull().any(axis=1)
x_full = x_full.loc[~ix,:]
```

```python
x_full
```

```python
from sklearn import linear_model
from sklearn.model_selection import KFold

kf = KFold(n_splits=5, random_state=None, shuffle=True)

bg = x_full['bg']
label = x_full['label']
xcols = ['diff5','diff10','diff15','diff20','diff25','diff30']
x = x_full[xcols]
y = x_full['difflabel']
#xcols = ['diff5','diff10','diff15','diff20','diff25','diff30','glycemicindex','calories','carbs','fiber','sugar','basal_insulin','bolus_insulin']

sum_rmsle = 0
for train_index, test_index in kf.split(x):
    lm = linear_model.LinearRegression()  
    model = lm.fit(x.iloc[train_index],y.iloc[train_index])
    predictions = lm.predict(x.iloc[test_index]) + np.log(bg.iloc[test_index].values)
    #predictions = np.log(bg.iloc[test_index].values)
    solution = pd.DataFrame(predictions, columns=['log_y'])
    sle = (predictions - np.log(label.iloc[test_index].values))**2
    msle = np.sum(sle)/len(test_index)
    rmsle = np.sqrt(msle)
    sum_rmsle += rmsle
    solution['real_y'] = np.log(label.iloc[test_index].values)
    solution['x'] = np.log(bg.iloc[test_index].values)
    solution.to_csv('c:/users/sunil/downloads/bgsolution.csv')

print(sum_rmsle)
```

```python
from sklearn.ensemble import RandomForestRegressor

bg = x_full['bg']
label = x_full['label']
#xcols = ['diff5','diff10','diff15','diff20','diff25','diff30']
xcols = ['diff5','diff10','diff15','diff20','diff25','diff30','glycemicindex','calories','carbs','fiber','sugar','basal_insulin','bolus_insulin']
x = x_full[xcols]
y = x_full['difflabel']

sum_rmsle = 0
for train_index, test_index in kf.split(x):
    rf = RandomForestRegressor()
    model = rf.fit(x.iloc[train_index],y.iloc[train_index])
    predictions = rf.predict(x.iloc[test_index]) + np.log(bg.iloc[test_index].values)
    #predictions = np.log(bg.iloc[test_index].values)
    solution = pd.DataFrame(predictions, columns=['log_y'])
    sle = (predictions - np.log(label.iloc[test_index].values))**2
    msle = np.sum(sle)/len(test_index)
    rmsle = np.sqrt(msle)
    sum_rmsle += rmsle
    solution['real_y'] = np.log(label.iloc[test_index].values)
    solution['x'] = np.log(bg.iloc[test_index].values)
    solution.to_csv('c:/users/sunil/downloads/bgsolution.csv')

print(sum_rmsle)
```
